```python
import pandas as pd
import openpyxl
import os
import re

# 读取Excel文件
df = pd.read_excel('../annotJournal.v4.xlsx')

journal_dafenqu = {}
journal_xiaofenqu = {}
journal_yujing = {}
journal_oa = {}
journal_leibie = {}
journal_dafenqu2022 = {}
journal_xiaofenqu2022 = {}
wenzs2023 = {}
feizyIF2023 = {}
ziyinlv2023 = {}
chubans2023 = {}

mylist = ['2018','2017','2016','2015','2014','2013','2012','2011','2010']
#mylist = ['2024','2023','2022','2021','2020','2019','1wmore']
#mylist = ['test']
#mylist = []

def get_text_before_paren(text):
    # 正则表达式匹配第一个括号或冒号前的字符
    match = re.search(r'(.*?)(\(|\:|\-)', text)
    if match:
        return match.group(1).strip()
    else:
        return text.strip()

def list_excel_files(directory):
    """
    遍历指定目录下所有以.xlsx结尾的文件，并返回文件路径列表。
    :return: 文件路径列表
    """
    return [os.path.join(dp, f) for dp, dn, fn in os.walk(os.path.expanduser(directory)) for f in fn if f.endswith('.xlsx')]
        
for index, row in df.iterrows():
    key = row[df.columns[0]].title()    
    key = get_text_before_paren(key)
    key = re.sub(r'^The ', "", key)
    key = re.sub(r'/', " ", key)
    key = re.sub(r'\. ', " ", key)
    key = re.sub(r'\, ', " ", key)
    journal_yujing[key] = str(row[df.columns[1]])
    journal_leibie[key] = str(row[df.columns[2]]) 
    journal_oa[key] = row[df.columns[3]]
    journal_dafenqu[key] = row[df.columns[4]]
    journal_xiaofenqu[key] = row[df.columns[5]]
    journal_dafenqu2022[key] = row[df.columns[6]]
    journal_xiaofenqu2022[key] = row[df.columns[7]]
    wenzs2023[key] = row[df.columns[8]]
    feizyIF2023[key] = row[df.columns[9]]
    ziyinlv2023[key] = row[df.columns[10]]
    chubans2023[key] = row[df.columns[11]]

for i in mylist:
    directory = './paper_donload/' + i + '/'
    excel_files = list_excel_files(directory)
    for path in excel_files:
        wb = openpyxl.load_workbook(path)
        ws = wb["Sheet"]

        for cur_row in range(2, ws.max_row+1):
            j_name = ws.cell(row=cur_row, column=6).value
            #j_short = ws.cell(row=cur_row, column=5).value
            try:
                j_name = j_name.title()
                j_name = get_text_before_paren(j_name)
                j_name = re.sub(r'^The ', "", j_name)
                j_name = re.sub(r'/', " ", j_name)
                j_name = re.sub(r'\. ', " ", j_name)
                j_name = re.sub(r'\, ', " ", j_name)
                #j_name = re.sub(r'arkinson\'s', "arkinsons", j_name)
                #j_name = re.sub(r'omen\'s', "omens", j_name)
                if j_name in journal_oa:
                    ws.cell(row=cur_row, column=16).value = journal_yujing[j_name]
                    ws.cell(row=cur_row, column=17).value = journal_leibie[j_name]
                    ws.cell(row=cur_row, column=18).value = journal_oa[j_name]
                    ws.cell(row=cur_row, column=19).value = journal_dafenqu[j_name]
                    ws.cell(row=cur_row, column=20).value = journal_xiaofenqu[j_name]
                    ws.cell(row=cur_row, column=21).value = journal_dafenqu2022[j_name]
                    ws.cell(row=cur_row, column=22).value = journal_xiaofenqu2022[j_name]
                    ws.cell(row=cur_row, column=23).value = wenzs2023[j_name]
                    ws.cell(row=cur_row, column=24).value = feizyIF2023[j_name]
                    ws.cell(row=cur_row, column=25).value = ziyinlv2023[j_name]
                    ws.cell(row=cur_row, column=26).value = chubans2023[j_name]
                    
                else:
                    ws.cell(row=cur_row, column=16).value = 'NA'
                    ws.cell(row=cur_row, column=17).value = 'NA'
                    ws.cell(row=cur_row, column=18).value = 'NA'
                    ws.cell(row=cur_row, column=19).value = 'NA'
                    ws.cell(row=cur_row, column=20).value = 'NA'
                    ws.cell(row=cur_row, column=21).value = 'NA'
                    ws.cell(row=cur_row, column=22).value = 'NA'
                    ws.cell(row=cur_row, column=23).value = 'NA'
                    ws.cell(row=cur_row, column=24).value = 'NA'
                    ws.cell(row=cur_row, column=25).value = 'NA'
                    ws.cell(row=cur_row, column=26).value = 'NA'
            except:     
                ws.cell(row=cur_row, column=16).value = 'NA'
                ws.cell(row=cur_row, column=17).value = 'NA'
                ws.cell(row=cur_row, column=18).value = 'NA'
                ws.cell(row=cur_row, column=19).value = 'NA'
                ws.cell(row=cur_row, column=20).value = 'NA'
                ws.cell(row=cur_row, column=21).value = 'NA'
                ws.cell(row=cur_row, column=22).value = 'NA'
                ws.cell(row=cur_row, column=23).value = 'NA'
                ws.cell(row=cur_row, column=24).value = 'NA'
                ws.cell(row=cur_row, column=25).value = 'NA'
                ws.cell(row=cur_row, column=26).value = 'NA'

        ws['P1'] = '是否预警'
        ws['Q1'] = 'SCI类别'
        ws['R1'] = 'OA信息'
        ws['S1'] = '中科院大类分区2023'
        ws['T1'] = '中科院小类分区2023'
        ws['U1'] = '中科院大类分区2022'
        ws['V1'] = '中科院小类分区2022'
        ws['W1'] = '文章数2023'
        ws['X1'] = '非自引IF2023'
        ws['Y1'] = '自引率2023'
        ws['Z1'] = '出版商'

        wb.save(path)
```
